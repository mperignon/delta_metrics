```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import cPickle as pickle

import clusterpy
import os
from glob import glob
```

```python
reduced_data = pd.read_pickle('ganges_metrics/metrics_results/reduced_data.p')
```

```python
params = reduced_data.columns.tolist()

filename = 'ganges_metrics/metrics_geosom/islands_properties_pca10_3'
c_data = clusterpy.importArcData(filename)


for n in range(8,12):
    
    directory = str(n) + 'x' + str(n)
    print directory
    
    output_dir = 'ganges_metrics/metrics_geosom/geoSom_PCA2/' + directory
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    nRows = n
    nCols = n

    c_data.cluster("geoSom", params, nRows=nRows, nCols=nCols, wType="queen", iters = 1000)
    c_data.exportArcData(output_dir + '/classes_' + directory)
    c_data.exportOutputs(output_dir + '/output_' + directory + ".csv")

    c_data.outputCluster = {'r2a': [], 'r2aRoot': []}
```
